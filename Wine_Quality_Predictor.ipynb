**From the above correlation metrix we find that â€˜total sulphur dioxideâ€™ and â€˜free sulphur dioxideâ€˜ are highly correlated features(0.667666	) so we will remove them.**

```python
df = df.drop(['total sulfur dioxide','free sulfur dioxide'], axis=1)
df.head()
```

```python
df['top quality'] = [1 if x > 5 else 0 for x in df.quality]
df.head()
```

```python
X = df.drop(['quality', 'top quality'], axis='columns')
Y = df['top quality']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=43)
```

```python
X_train.shape, X_test.shape
```

# Ensemble Learning
XGBClassifier

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn import metrics
print('Training Accuracy : ', metrics.roc_auc_score(y_train, model.predict(X_train)))
```

```python
print('Validation Accuracy : ',metrics.roc_auc_score(y_test, model.predict(X_test)))
```

```python
print(metrics.classification_report(y_test, model.predict(X_test)))
```

# Ensemble Learning
RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
error_accuracy = metrics.r2_score(y_pred,y_test)
test_accuracy = metrics.accuracy_score(y_pred,y_test)
```

```python
print('Test Accuracy : ',test_accuracy)
```
